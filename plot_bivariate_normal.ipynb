```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def plot_bivariate_normal(covmat, filename, x_range=(-4, 4), y_range=(-4, 4), grid_size=100):
    """
    Plots the bivariate normal distribution in both 3D and top view.

    Parameters:
    - covmat: Covariance matrix for the bivariate normal distribution.
    - filename: Base filename for saving the plots.
    - x_range: Tuple specifying the range for the x-axis.
    - y_range: Tuple specifying the range for the y-axis.
    - grid_size: Number of points in each dimension for the grid.
    """
    # Mean vector
    mean = [0.0, 0.0]
    
    # Create a grid of points
    x = np.linspace(x_range[0], x_range[1], grid_size)
    y = np.linspace(y_range[0], y_range[1], grid_size)
    X, Y = np.meshgrid(x, y)
    
    # Stack the grid points
    pos = np.dstack((X, Y))
    
    # Evaluate the PDF
    Z = multivariate_normal.pdf(pos, mean=mean, cov=covmat)
    
    # Create a figure with subplots
    fig = plt.figure(figsize=(14, 10))
    
    # Format the covariance matrix for the title
    cov_str = f"Covariance Matrix: {covmat}"
    
    # 3D view
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_surface(X, Y, Z, cmap='plasma', edgecolor='black', alpha=1)
    ax1.set_xlabel('X', fontsize=14)
    ax1.set_ylabel('Y', fontsize=14)
    ax1.set_zlabel('Probability Density', fontsize=14)
    ax1.set_title(f'3D View of Bivariate Normal Distribution\n{cov_str}', fontsize=16)
    
    # Top view
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.plot_surface(X, Y, Z, cmap='plasma', edgecolor='black', alpha=1)
    ax2.view_init(elev=90, azim=0)
    ax2.set_xlabel('X', fontsize=14)
    ax2.set_ylabel('Y', fontsize=14)
    ax2.set_title(f'Top View of Bivariate Normal Distribution\n{cov_str}', fontsize=16)
    
    # Remove ticks
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.set_zticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.set_zticks([])
    
    # Adjust layout to prevent cutting off labels
    plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)
    
    # Save the figure
    #with open(f'./../pics/{filename}.png', 'wb') as f:
    #    plt.savefig(f, transparent=True)
    
    plt.show()

# Example usage
covariance_matrix = [[1, 0.5], [0.5, 1]]  # Example covariance matrix
plot_bivariate_normal(covariance_matrix, 'bivariate_normal_distribution')

```
